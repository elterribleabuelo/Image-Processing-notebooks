# Ejercicio 1

Segmentar de los dos patos. La salida debe ser la imagen original con un borde amarillo rectangular que encierre cada pato.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
os.chdir(r"C:\Users\titos\Github\Procesamiento Imagenes\Cap01_Introduccion\imagenes")
```

```python
IMG01 = cv2.imread("./IMG01.png")
IMG01 = cv2.cvtColor(IMG01,cv2.COLOR_BGR2RGB)
plt.figure(1,figsize=(8,10))
plt.imshow(IMG01)
plt.show()
```

```python
rojo = IMG01[:,:,0]
verde = IMG01[:,:,1]
azul = IMG01[:,:,2]
plt.figure(1,figsize=(20,18))
plt.subplot(1,3,1)
plt.title("Banda roja")
plt.imshow(rojo,cmap='gray')
plt.subplot(1,3,2)
plt.title("Banda verde")
plt.imshow(verde,cmap='gray')
plt.subplot(1,3,3)
plt.title("Banda azul")
plt.imshow(azul,cmap='gray')
plt.show()
```

```python
def SegmentadorPatos(IMG01):
    aux = np.zeros((IMG01.shape[0],IMG01.shape[1]))
    
    # Umbralización según canal
    for i in range(IMG01.shape[0]):
        for j in range(IMG01.shape[1]):
            if (IMG01[i,j,2] < 65) or (IMG01[i,j,0] > 180 and 100 < IMG01[i,j,1] < 170)  :
                aux[i,j] = 1
                
    # Operaciones morfologicas
    kernel = np.ones((3,3))
    patos = cv2.morphologyEx(aux,cv2.MORPH_OPEN,kernel)
    patos = cv2.dilate(aux,kernel,2)
    patos = patos.astype('uint8')
    
    # Contornos
    cnts_patos,_ = cv2.findContours(patos,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    IMG_Output = IMG01.copy()

    # Dibujamos contornos
    for c in cnts_patos:
        area = cv2.contourArea(c)
        x,y,w,h = cv2.boundingRect(c)
        if area > 450:
            cv2.rectangle(IMG_Output,(x,y) , (x+w,y+h), (255,255,0),2)
    plt.figure(1,figsize=(20,10))
    plt.subplot(1,2,1)
    plt.title("Imagen segmentada")
    plt.imshow(patos,cmap='gray')
    plt.subplot(1,2,2)
    plt.title("Output")
    plt.imshow(IMG_Output)
```

```python
SegmentadorPatos(IMG01)
```

# Ejercicio 2

Cacule la longitud del lapiz en pixles (opcional: medir el angulo que hace el lapiz con una linea horizontal).

```python
IMG02 = cv2.imread("./IMG02.png")
aux = IMG02.copy()
IMG02 = cv2.cvtColor(IMG02,cv2.COLOR_BGR2GRAY)
umbralizado = cv2.threshold(IMG02,55,255,cv2.THRESH_BINARY_INV)[1]
umbralizado[0:10,:] = 0
plt.figure(1,figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(IMG02,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(umbralizado,cmap='gray')
plt.show()
print(IMG02.shape)
```

```python
# Punto 1 y 2 (izquierda)
count1 = 0
count2 = 0

## Buscamos por fila
for i in range(IMG02.shape[0]): # filas
    for j in range(IMG02.shape[1]): # columnas
        if umbralizado[i][j] == 255:
            count1 = count1 + 1
        if count1 == 1:
            P1 = (i,j)

## Buscamos por columna
for j in range(IMG02.shape[1]): # columnas
    for i in range(IMG02.shape[0]): # filas
        if umbralizado[i][j] == 255:
            count2 = count2 + 1
        if count2 == 1:
            P2 = (i,j)
print(P1)
print(P2)
```

```python
# Punto 3 y 4 (derecha)
count3 = 0
count4 = 0

## Buscamos por fila
for i in range(IMG02.shape[0]): # filas
    for j in range(IMG02.shape[1]): # columnas
        if umbralizado[IMG02.shape[0] -1 - i][j] == 255:
            count3 = count3 + 1
        if count3 == 1:
            P3 = (IMG02.shape[0] -1 - i,j)

## Buscamos por columna
for j in range(IMG02.shape[1]): # columnas
    for i in range(IMG02.shape[0]): # filas
        if umbralizado[i][IMG02.shape[1] - 1 - j] == 255:
            count4 = count4 + 1
        if count4 == 1:
            P4 = (i,IMG02.shape[1] - 1 - j)
print(P3)
print(P4)
```

```python
m = cv2.moments(umbralizado)
x_c = m['m10']/m['m00']
y_c = m['m01']/m['m00']
print('x=',x_c, 'y=',y_c)
```

```python
# Visualizamos estos puntos
#cv2.circle(aux,(P1[1],P1[0]),3,(2555,0,0),-1)
cv2.circle(aux,(P2[1],P2[0]),3,(255,0,0),-1)
#cv2.circle(aux,(P3[1],P3[0]),3,(255,0,0),-1)
cv2.circle(aux,(P4[1],P4[0]),3,(255,0,0),-1)
cv2.circle(aux,(int(x_c),int(y_c)),3,(255,0,0),-1)
cv2.line(aux,(P2[1],P2[0]),(int(x_c),int(y_c)),(0,255,0),1)
cv2.line(aux,(P4[1],P4[0]),(int(x_c),int(y_c)),(0,255,0),1)
plt.figure(1,figsize=(20,16))
plt.imshow(aux)
plt.show()
```

```python
# Distancia
(x1,y1) = (P2[1],P2[0])
distancia_1 = (x_c-x1)**2 + (y_c-y1)**2
distancia_1 = np.sqrt(distancia_1)

# Tangente del angulo formado por las 2 rectas verdes
m2 = (y_c - y1)/(x_c - x1)
m1 = (P4[0] - y_c)/(P4[1] - x_c)

# Angulo en radianes
tang_alpha = np.abs((m2-m1)/(1+m2*m1))
alpha = np.arctan(tang_alpha) # radianes

# Distancias
distancia_pitagorica = np.sqrt((x_c - P4[1])**2 + (y_c - P4[0])**2)
distancia_2 = distancia_pitagorica*np.cos(alpha)

longitud_lapiz = distancia_1 + distancia_2
print("La longitud del lapiz es: {} pixeles".format(longitud_lapiz))
```
