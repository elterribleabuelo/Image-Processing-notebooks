<h1 align="center"><strong>Transformada de Wavelets</strong></h1>

Las wavelets son funciones en el conjunto de los números reales para el conjunto de los números reales, cada uno de los cuales deriva de una función madre mediante traslación y escala.

En la transfrmada de wavelet las funciones base son funciones que se definen en una ubicación y también tienen que ver con la frecuencia. por lo tanto para analizar con precisión las señales que tienen cambios abruptos , necesitamos utilizar una nueva clase de funciones que esten bien localizadas en el tiempo y la frecuencia, esto nos lleva al tema de las wavelets.

La primera y más básica wavelet es la Haar, la transformada de Haar tiene como funciones base a funciones que toman los valores de -1 , 0 y 1 , estas funciones base tienen la misma estructura pero ubicadas en distintas partes.

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from math import cos,sin,sqrt,pi
import pywt
os.chdir(r"C:\Users\titos\Github\Procesamiento Imagenes\Cap05_Mejoramiento_Frecuencia\imagenes")
```

```python
# Ejemplo 1: Hallando los bordes de distinta orientación

# Imagen de entrada
lenna = cv2.imread("./Lenna.png",0) # gris

# Definimos el wavelet
wavelet = pywt.Wavelet('haar')

# Applicando wavelet a lenna
lw = pywt.dwt2(lenna, wavelet, mode='per')

print("1. Número de niveles:",len(lw) - 1)


fig = plt.figure(1,figsize=(12,12))
st = fig.suptitle("Fig 1: Subbandas", fontsize="x-large")

ax1 = fig.add_subplot(2,2,1)
ax1.set_title("LL")
#plt.axis('off')
ax1.imshow(lw[0], cmap='gray') # Coeficientes de aproximacion

ax2 = fig.add_subplot(2,2,2)
ax2.set_title("LH")
#plt.axis('off')
ax2.imshow(lw[1][0], cmap='gray') # detalle horizontal

ax3 = fig.add_subplot(2,2,3)
ax3.set_title("HL")
#plt.axis('off')
ax3.imshow(lw[1][1], cmap='gray') # detalle vertical

ax4 = fig.add_subplot(2,2,4)
ax4.set_title("HH")
#plt.axis('off')
ax4.imshow(lw[1][2], cmap='gray') # detalle diagonal

fig.tight_layout()
```

De a Figura anterior, LL es una aproximación suavizada de la imagen de Lenna y las otras tres matrices contienen coeficientes de ondas que capturan oscilaciones de alta frecuencia en direcciones vertical(LH), horizontal(HL) y diagonal(HH).

```python
# Ejemplo 2: Compresión de imágenes

wavelet = pywt.Wavelet('haar')
ex = cv2.imread("./Lenna.png",0)

# Aplicamos Haar a la imagen de entrada
WaveletCoeffs = pywt.wavedec2(ex,wavelet)
print("Numero de subbandas encontradas:",len(WaveletCoeffs))

# Eliminamos las altas frecuencias
subbandas = 2
new_image = pywt.waverec2(WaveletCoeffs[:-subbandas], wavelet) # Usamos todos los coeficientes menos los dos ultimos
print("Numero de subbandas usadas:",len(WaveletCoeffs[:-subbandas]))


plt.figure(1,figsize=(16,16))
plt.subplot(1,2,1)
plt.title("Imagen de entrada")
plt.imshow(ex,cmap='gray')

plt.subplot(1,2,2)
plt.title("Imagen comprimida")
plt.imshow(new_image,cmap='gray')
plt.show()
```
