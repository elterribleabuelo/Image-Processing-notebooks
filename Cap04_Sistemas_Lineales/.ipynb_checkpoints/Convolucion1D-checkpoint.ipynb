# 1.Función impulso

**¿Qué es la función impulso?**

La función impulso también conocida como la función de Dirac Delta, tiene la siguiente regla de correspondencia:

$$\int_{-\infty}^{+\infty} \delta(x)\,dx = 1$$

Cuando: $$x = 0 \text{ ; }  \delta(x) = {\infty} \$$

$$x \neq 0 \text{ ; }  \delta(x) = {0} \$$

Propiedades

1. $$\delta(x)\ =  \ lim_{a \to 0}(\frac{1}{a} \prod \frac{x}{a}) $$

2. $$\int_{-\infty}^{+\infty} \text{A} \delta(x)\,dx = A$$

3. $$\int_{-\infty}^{+\infty} f(x) \delta(x)\,dx = f(x)$$

4. $$ y(t) = \delta(t)\ \ast y(t)  $$


**¿Para que sirve la función impulso?**

De la propiedad 3, esta es usada en el muestreo de una señal , lo que se hace es multiplicar una señal f(x) por un tren de impulsos a una frecuencia dada.

El resultado proveniente de la propiedad 4 es muy importante ya que permite averiguar lo que está pasando en una caja negra, si es un sistema lineal se manda un impulso el cual tiene un valor grande en algun lado y despues baja, por lo que se puede saber que es lo que pasa en ese sistema lineal.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from IPython.display import Image
from IPython.core.display import HTML
```

```python
# INGRESO
u = lambda t: np.piecewise(t,t>=0,[1,0])

a = -10
b = 10
dt = 0.1

# PROCEDIMIENTO
t = np.arange(a, b, dt)
u0 = u(t)
udt = u(t-dt)
impulso = u0 - udt

# SALIDA - GRAFICA
plt.figure(4)
plt.plot(t,impulso)

plt.margins(dt)
plt.xlabel('t')
plt.ylabel('impulso()')
plt.grid()
plt.show()
```

# 2.Convolución 1D

## 2.1. Filtrado simple con una ventana promedio

```python
n = 1000
i = np.array(range(n))
t = i*8*np.pi/n
x = np.sin(t)+np.sin(2*t)

r  = np.random.random_sample(size = n)-0.5

xr  = x + r
plt.figure(figsize=(20,9))
plt.plot(t,xr) # Señal de alta frecuencia
plt.plot(t,x)# Señal de baja frecuencia
plt.legend(['xr','x'])
plt.show()
```

```python
m = 15
g = np.ones((m,))/m # filtro

y = np.convolve(xr,g,'same')

plt.figure(figsize=(20,9))
plt.plot(t,xr) # señal con ruido
#plt.plot(t,x)
plt.plot(t,y) # señal filtrada
plt.legend(['xr','y'])
plt.show()
```

## 2.2.Separación de dos señales (con Gaussianas) 

* $x_1$: primera señal de frecuencia baja

* $x_2$: primera señal de frecuencia alta

* $x = x_1 + x_2$

* $z = x_1 + x_2 + $ ruido

Problema: a partir de $z$ encontrar $x_1$ y $x_2$

Solución:

$y_1 = $ filtro pasa (muy) bajo de $z$ con una Gaussiana $h_1$ de ancho 7: aquí solo pasa la sinusoide de frecuencia baja

$y_2 = $ filtro pasa bajo de $z$ con una Gaussiana $h_1$ de ancho 1: aquí se elimina el ruido, $y_2$ es parecido a $x$.

La reconstrucción de $x_1$ es $y_1$.

La reconstrucción de $x_2$ es $y_2 - y_1$.

: primera señal de frecuencia baja

```python
Image(filename = "../Cap04_Sistemas_Lineales/imagenes/senal.png", width=1000, height=1000)
```

```python
n  = 1000
i  = np.array(range(n))
t  = i*8*np.pi/n
x1 = np.sin(t)
x2 = 0.3*np.sin(20*t)
x  = x1+x2
r  = 0.2*(np.random.random_sample(size = n)-0.5)
z  = x + r
plt.figure(figsize=(20,9))
plt.plot(t,x1) # baja frecuencia
plt.plot(t,x2) # alta frecuencia
plt.plot(t,z) # sumatoria de ambas
plt.legend(['x1','x2','z'])
plt.show()
```

```python
m = 25
h1 = cv.getGaussianKernel(m, 7)
h2 = cv.getGaussianKernel(m, 1)
plt.plot(h1)
plt.plot(h2)
plt.legend(['h1','h2'])
plt.show()
```

```python
y1 = np.convolve(z,h1[:,0],'same') # Gaussiana más ancha filtra las frecuencias mas bajas
y2 = np.convolve(z,h2[:,0],'same') # Gaussiana menos ancha filtra las frecuencias mas altas
plt.figure(figsize=(20,9))
plt.plot(y2)
plt.plot(y1)
plt.legend(['y2','y1'])
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.plot(x1) # señal original
plt.plot(y1) # señal reconstruida
plt.plot(x1-y1) # error
plt.legend(['x1','y1','error'])
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.plot(x2) # señal original
plt.plot(-y1+y2) # señal reconstruida
plt.plot(x2-(-y1+y2)) # error
plt.legend(['x2','y2-y1','error'])
plt.show()
```

```python
# El error se produce porque el ruido tambien tiene componentes en la frecuencias más bajas.
```

## 3) Simulación de una imagen con movimeinto horizontal

```python
X = cv.imread('../Cap04_Sistemas_Lineales/imagenes/zebra_bw.png',0)
plt.figure(1,figsize=(10,10))
plt.imshow(X,cmap = 'gray')
```

```python
# Convolución línea por línea
m = 25
g = np.ones((m,))/m
(N,M) = X.shape
Y = np.zeros((N,M))
for i in range(N):
    Y[i,:] = np.convolve(X[i,:],g,'same')
plt.figure(1,figsize=(10,10))
plt.imshow(Y,cmap='gray')
```

```python
# Convolución de toda la imagen 
h = g.reshape(1,m) # vector fila
Y = cv.filter2D(X,-1,h)
plt.figure(1,figsize=(10,10))
plt.imshow(Y,cmap='gray')
```

## 4) Detección de bordes

```python
Xd = X.astype(float)
th = 50
h1 = np.array([-1,0,1])
Y1 = np.abs(cv.filter2D(Xd,-1,h1))
Z1 = Y1>th
h2 = h1.reshape(1,3)
Y2 = np.abs(cv.filter2D(Xd,-1,h2))
Z2 = Y2>th
Z  = np.logical_or(Z1,Z2)
plt.figure(1,figsize=(10,10))
plt.imshow(Z*255,cmap='gray')
```

```python
(N,M) = X.shape
D = np.zeros((N,M,3))
D[:,:,1] = np.multiply(X,(1-Z))
D[:,:,0] = np.multiply(X,(1-Z))
D[:,:,2] = X
plt.figure(1,figsize=(10,10))
plt.imshow(D.astype('uint8'),cmap='gray')
```

```python
T1 = np.sqrt(np.multiply(Y1,Y1)+np.multiply(Y2,Y2))
T1 = np.log(T1+1)
T1 = T1/np.max(T1)*255
plt.figure(1,figsize=(10,10))
plt.imshow(255-T1,cmap='gray')
```
