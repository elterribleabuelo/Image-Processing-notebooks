# 1.Problema 01

## 1.1.Enunciado

En este problema se desea segmentar el sol y la gaviota de la imagen atardecer.png.

Para la solución se debe usar operaciones morfológicas. Se recomienda usar un filtro mediana para eliminar la gaviota, y una erosión para eliminar el sol. La segmentación se obtiene mediante restas de la imagen original con las imagenes filtradas y un umbral adecuado para la gaviota y otro para el sol. Soluciones que no consideren operaciones morfológicas no serán consideras.

## 1.2.Solución

```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage, misc
```

```python
os.chdir(r"C:\Users\titos\Github\Procesamiento Imagenes\Cap07_Morfología\imagenes")
```

```python
%ls
```

```python
def Normalizar(f):
    faux = np.ravel(f).astype(float)
    minimum = faux.min()
    maximum = faux.max()
    g = (faux-minimum)*(255) / (maximum-minimum)
    r = g.reshape(f.shape).astype(np.uint8)
    return(r)
```

```python
def f1(image):
    I1 = cv2.imread(image)
    I2 = I1.copy()
    
    ### Eliminamos la gaviota ###
    
    # Filtro mediana
    median = ndimage.median_filter(I1, size = 7) # size = 9
    
    ### Eliminamos el sol ###
    kernel = np.ones((9, 9), np.uint8) # (9,9)
    erosion = cv2.erode(median, kernel, iterations = 5) # iterations = 3 0 5
    
    # Segmentamos la gaviota y sol
    resta1 =  Normalizar(I1 - erosion)
    umbral1 = cv2.threshold(resta1,195,255,cv2.THRESH_BINARY)[1]
    umbral2 = cv2.threshold(resta1,100,255,cv2.THRESH_BINARY)[1]
    
    # Contornos
    cnts_gaviota,_ = cv2.findContours(umbral1[:,:,0],cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    cnts_sol,_ = cv2.findContours(umbral2[:,:,0],cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    
    # Dibujamos contornos
    for c1,c2 in zip(cnts_gaviota,cnts_sol):
        area1 = cv2.contourArea(c1)
        area2 = cv2.contourArea(c2)
        x1,y1,w1,h1 = cv2.boundingRect(c1)
        x2,y2,w2,h2 = cv2.boundingRect(c2)
        if 27 < area1 < 28 or 500 < area2 < 1000:
            cv2.drawContours(I2, [c1], 0, (255,0,0),1)
            cv2.drawContours(I2, [c2], 0, (255,0,0),1)
    plt.figure(1,figsize=(12,12))
    plt.imshow(I2,cmap='gray')
    plt.show()
    
    return 
```

```python
f1("atardecer.png")
```

# 2.Problema 02

## 2.1.Enunciado

En este problema se debe realizar un gradiente morfológico de la imagen tazmania.png. El output debe ser una imagen en la que se muestre el valor abosluto del gradiente morfológico.


[opcional]Realizar el gradiente con operadores lineales y comparar (visualmente) el gradiente obtenido con el gradiente morfológico. El gradiente con operador lineal se calcula usando dos máscaras: [-1 0 1] para obtener Gx, el gradiente en x; [-1 0 1]' para obtener Gy, el gradiente en y. El gradiente de la imagen puede calcularse como la magnitud de ambos gradientes: sqrt(Gx^2+Gy^2).

## 2.2.Solución

```python
tazmania = cv2.imread('tazmania.png')
plt.figure(1,figsize=(10,12))
plt.imshow(tazmania,cmap='gray')
plt.show()
```

```python
kernel_tazmania = np.ones((5, 5), np.uint8) # (9,9)
# dilatacion
d = cv2.dilate(tazmania,kernel_tazmania,2)
# erosion
e = cv2.erode(tazmania,kernel_tazmania,2)
# resta
resta = d- e
plt.figure(1,figsize=(10,12))
plt.imshow(resta,cmap='gray')
plt.show()
```

```python
# Máscaras
m_X = np.array([[-1,0,1]])
m_Y = m_X.T

# Gradiente en X
G_X = cv2.filter2D(tazmania,-1,m_X) # bordes verticales

# Gradiente en Y
G_Y = cv2.filter2D(tazmania,-1,m_Y) # bordes horizontales

# Gradiente resultante
suma = np.power(G_X,2) +  np.power(G_Y,2)
G = np.sqrt(suma)
G = np.array(G,dtype=np.uint8)
G = Normalizar(G)

# Graficamos
plt.figure(1,figsize=(30,30))

plt.subplot(1,3,1)
plt.title("Gradiente en X",fontsize = 18)
plt.imshow(G_X,cmap='gray')
plt.subplot(1,3,2)
plt.title("Gradiente en Y",fontsize = 18)
plt.imshow(G_Y,cmap='%echoray')
plt.subplot(1,3,3)
plt.title("Gradiente total",fontsize = 18)
plt.imshow(G,cmap='gray')

plt.show()
```

```python
# Comparación
plt.figure(1,figsize=(30,30))
plt.subplot(1,3,1)
plt.title("Gradiente morfológico",fontsize = 18)
plt.imshow(resta,cmap='gray')

plt.subplot(1,3,2)
plt.title("Gradiente con operadores lineales",fontsize = 18)
plt.imshow(G,cmap='gray')
plt.show()
```

# 3.Problema 03

## 3.1.Enunciado

En este problema se desea segmentar las gotas medias y grandes de la imagen gotas.png.El output debe ser una imagen en la que se muestre la segmentación(no importa que la gota no sea segmentada completamente).

Para la solución se debe usar operaciones morfológicas. Se recomienda usar un filtro TopHat para eliminar la iluminación irregular, luego usar un umbral para detectar las zonas más claras y por último un filtro morfológico que elimine las regiones muy pequeñas. Soluciones que no consideren operaciones morfológicas no serán consideras.

## 3.2.Solución

```python
gotas = cv2.imread('gotas.png')
dibujar = gotas.copy()

plt.figure(1,figsize=(10,12))
plt.title("Imagen original",fontsize = 18)
plt.imshow(gotas,cmap='gray')
plt.show()
```

```python
# Resaltar elementos blancos en la imagen

kernel_tophat = cv2.getStructuringElement(cv2.MORPH_RECT,(27,27)) #(17,17)
tophat = cv2.morphologyEx(gotas, cv2.MORPH_TOPHAT, kernel_tophat,5) # 5
tophat = Normalizar(tophat)

plt.figure(1,figsize=(10,10))
plt.title("Imagen con TopHat",fontsize = 18)
plt.imshow(tophat,cmap='gray')
plt.show()
```

```python
# Segmentación por umbralización
img_umbralizada = cv2.threshold(tophat,100,255,cv2.THRESH_BINARY)[1]
plt.figure(1,figsize=(10,10))
plt.title("Imagen umbralizada",fontsize = 18)
plt.imshow(img_umbralizada,cmap='gray')
plt.show()
```

```python
kernel_erode = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(11,11))
img_umbralizada = cv2.morphologyEx(img_umbralizada,cv2.MORPH_OPEN,kernel_erode,3)                  

plt.figure(1,figsize=(10,10))
plt.title("Imagen umbralizada + erosion",fontsize = 18)
plt.imshow(img_umbralizada,cmap='gray')
plt.show()
```

```python
# Contornos
cnts_gotas,_ = cv2.findContours(img_umbralizada[:,:,0],cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

# Dibujamos contornos
for c1 in cnts_gotas:
    area = cv2.contourArea(c1)
    if area>280:
        cv2.drawContours(dibujar, [c1], 0, (255,0,0),2)

plt.figure(1,figsize=(10,10))
plt.title("Imagen segmentada",fontsize = 18)
plt.imshow(dibujar,cmap='gray')
plt.show()
```
