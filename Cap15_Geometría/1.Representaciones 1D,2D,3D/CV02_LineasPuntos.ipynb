# 1. Puntos y líneas 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Producto Punto

```python
m1 = [10,20,1] #(x1,y1) = (10,20)
m2 = [20,30,1] #(x2,y2) = (20,30)
print(np.dot(m1,m2))
```

```python
m1 = [10,20,1]
m2 = [20,30,1]
print(np.cross(m1,m2))
```

## Gráfica de Puntos y Líneas

```python
m1 = [10,20,1]
m2 = [20,30,1]

xs = [m1[0], m2[0]]
ys = [m1[1], m2[1]]

plt.text(xs[0]-.2, ys[0]+1, '$m_1$')
plt.plot(xs[0], ys[0], marker='o', markerfacecolor='red', markersize=12)

plt.text(xs[1]-.3, ys[1]-1, '$m_2$')
plt.plot(xs[1], ys[1], marker='o', markerfacecolor='orange', markersize=12)

plt.plot(xs,ys,color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Línea entre m1 y m2')
plt.show()
```

# 2. Ecuación de una recta a partir de dos puntos

Si el primer punto 2D en coordenadas homogéneas es ${\bf m}_1 = [x_1 \ y_1 \ 1]^{\sf T}$, y el segundo es ${\bf m}_2 = [x_2 \ y_2 \ 1]^{\sf T}$, entonces la definición de la recta que pasa por ambos puntos es:

$\ell = {\bf m}_1 \times {\bf m}_2 = [a \ b \ c]^{\sf T}$.

Es decir, la ecuación de la recta es:

$ax + by + c = 0$

```python
m1 = [10,20,1]
m2 = [20,30,1]

ell = np.cross(m1,m2) # <==== recta que pasa por ambos puntos
a   = ell[0]
b   = ell[1]
c   = ell[2]

print('ell      = ',ell)
print('(a,b,c) = ',(a,b,c))

# a*x + b*y + c = 0
# y = -(a*x+c)/b
```

## Evaluación de la ecuación de una recta en x = 14

```python
m1 = [10,20,1]
m2 = [20,30,1]

xs = [m1[0], m2[0]]
ys = [m1[1], m2[1]]


ell = np.cross(m1,m2) # <==== recta que pasa por ambos puntos
a   = ell[0]
b   = ell[1]
c   = ell[2]

plt.plot(xs,ys,color='blue')
plt.xlabel('x')
plt.ylabel('y')

x = 14
y = -(a*x+c)/b

plt.plot(x, y, marker='o', markerfacecolor='red', markersize=12)

print('(x,y)=',(x,y))
```

# Intersección de dos rectas

Si la primera recta 2D en coordenadas homogéneas es ${\bf \ell}_1 = [a_1 \ b_1 \ c_1]^{\sf T}$, y la segunda es ${\bf \ell}_2 = [a_2 \ b_2 \ c_2]^{\sf T}$, entonces la definición de la intersección que pasa por ambas rectas es:

${\bf m}_i = {\bf \ell}_1 \times {\bf \ell}_2 = [p \ q \ r]^{\sf T}$.

Es decir, el punto es:

$x_i = \frac{p}{r}$

$y_i = \frac{q}{r}$


## Definición de recta 1

```python
ell_1 = [1, -1, 10] # la recta del ejercicio anterior
```

## Definición de recta 2

```python
ell_2 = [1,1,-40]
```

## Gráfica de ambas rectas

```python
# Recta 1
a1   = ell_1[0]
b1   = ell_1[1]
c1   = ell_1[2]

x1 = np.array([10,20])
y1 = -(a1*x1+c1)/b1

plt.plot(x1,y1,color='blue')

# Recta 2
a2   = ell_2[0]
b2   = ell_2[1]
c2   = ell_2[2]

x2 = np.array([10,20])
y2 = -(a2*x2+c2)/b2

# Gráfica
plt.plot(x2,y2,color='green')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

## Intersección

```python
mi = np.cross(ell_1,ell_2) # <==== intersección de ambas rectas

xi = mi[0]/mi[2]
yi = mi[1]/mi[2]
print('(xi,yi) = ',(xi,yi))


plt.plot(x1,y1,color='blue')
plt.plot(x2,y2,color='green')
plt.plot(xi, yi, marker='o', markerfacecolor='red', markersize=12)

plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
